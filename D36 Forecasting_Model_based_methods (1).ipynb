### Forecating different models

Additive seasonality Quadric trend

Linear Model: Yt = B. + Bit + 

Exponential Model: Log (Y) = B. + Bt + 

Quadratic Model: Yt = Bo + Bt + B2t**2 + 

Additive Seasonality: Y1 = B. + B1DJan + B2DFeb + B3DMar + ...... + B11DNov + E


Additive Seasonality with Quadratic Trend:
Yt = Bo + Bit + B2t2 + B3DJan + B4DFeb + B5DMar + ...... + B13DNov

Multiplicative Seasonality:
Log (Y) = B. + B1DJan + B2DFeb + B3DMar + .. + B11DNOV

```python
# Linear Model
import statsmodels.formula.api as smf

linear_model = smf.ols('Footfalls~t',data=Train).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(Test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(pred_linear))**2))
rmse_linear

# First fitting OLS --> linear regression
# Footfalls -- InDependent  , t -- Dependent ,data=Train

# squareroot[mean(orginal-predit)**2]
# OLS Accuracy = 209.92559
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
l1 = LinearRegression()
l1.fit(Train['t'].values.reshape(-1, 1),Train['Footfalls'].values.reshape(-1, 1))
ypred = l1.predict(Test['t'].values.reshape(-1, 1))
print(np.sqrt(mean_squared_error(Test['Footfalls'].values.reshape(-1, 1),ypred)))

# Using Linear regression By default
# fit --> Footballs data
# reshape with some values
# OLS and Liner regression both are same
```

```python
#Exponential

Exp = smf.ols('log_footfalls~t',data=Train).fit()
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(Test['t'])))
rmse_Exp = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(np.exp(pred_Exp)))**2))
rmse_Exp


# Apply log_footfalls (data of)  t
# log_footfalls data on comparsion with time(t)
# error is increase
```

```python
#Quadratic

Quad = smf.ols('Footfalls~t+t_square',data=Train).fit()
pred_Quad = pd.Series(Quad.predict(Test[["t","t_square"]]))
rmse_Quad = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(pred_Quad))**2))
rmse_Quad

# predit test data t,t_square
```

```python
#Additive seasonality

add_sea = smf.ols('Footfalls~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data=Train).fit()
pred_add_sea = pd.Series(add_sea.predict(Test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']]))
rmse_add_sea = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(pred_add_sea))**2))
rmse_add_sea

# y variable with all the seasons (jan, feb,...dec)
# predict
```

```python
# Additive Seasonality Quadratic

add_sea_Quad = smf.ols('Footfalls~t+t_square+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data=Train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(Test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','t','t_square']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad

# Football data is preticted using (t+t_square+jan+feb+..+dec)
# predict
# Large dif
# It gives very least error
# it is good ASQ
```

```python
##Multiplicative Seasonality

Mul_sea = smf.ols('log_footfalls~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data = Train).fit()
pred_Mult_sea = pd.Series(Mul_sea.predict(Test))
rmse_Mult_sea = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea

# y --> log_footfalls InDependent
# t --> Dependent
# aditive sea an mult sea are very simmular formula
```

```python
#Multiplicative Additive Seasonality

Mul_Add_sea = smf.ols('log_footfalls~t+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data = Train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(Test))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea

# log_footfalls --> Dependent set
# t and all the seasons
```

```python
# Multiplicative quadratic Seasonality

Mul_Add_sea = smf.ols('log_footfalls~t+t_square+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data = Train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(Test))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(Test['Footfalls'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea

```
